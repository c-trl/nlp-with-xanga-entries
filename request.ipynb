#Pulling text from WordPress with Requests

```python
url = 'http://crkyo.wordpress.com/page/'
pages = range(1,26)

urls = []

for x in pages:
    urls.append(url+str(x))  
```

```python
urls[24]
```

```python
uhtml = []

for x in urls:
    r = requests.get(x)
    uhtml.append(r.text)
```

```python
type(uhtml), type(uhtml[24])
```

#Unicode to ASCII Conversion

```python
source = []

for upage in uhtml:
    page = upage.encode('utf-8').strip()
    source.append(page)
```
