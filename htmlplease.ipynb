```python
import time
import urllib2
from urllib2 import urlopen
import datetime
import pandas as pd
```

```python
url = 'http://crkyo.wordpress.com/page/'
thirty = range(1,31)

for x in thirty:
    urls = list()
    urls.append(url+str(x))  
```

```python
def reader():
    for x in urls:
        try:
            print urllib2.urlopen(x).read()
        except urllib2.HTTPError, e:
            print e.code
            print e.msg
            print e.headers
            print e.fp.read()
            
reader()
```

```python
html = str(reader())
```

```python
def find_between(html, first, last ):
    try:
        start = html.index( first ) + len( first )
        end = html.index( last, start )
        return html[start:end]
    except ValueError:
        return "oops!"
    
print find_between(html, "<div class=\"entry-content\">", "<!-- .entry-content -->" )
```
